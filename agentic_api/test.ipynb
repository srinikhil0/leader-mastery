```python
import chromadb

# Connect to ChromaDB
client = chromadb.PersistentClient(path="1_db")

# List available collections
collections = client.list_collections()
print("Available collections:")
for col in collections:
    print(" -", col)

# Choose your collection
collection_name = "Real-Estat"
collection = client.get_collection(name=collection_name)

# Fetch all documents (limit is optional)
results = collection.get(include=["documents", "metadatas", "embeddings"])

# Display documents
print("\n📄 Stored Documents in Collection:")
for idx, doc in enumerate(results["documents"]):
    print(f"\n--- Document {idx+1} ---")
    print("ID:", results["ids"][idx])  # IDs are always returned, even if not included
    print("Metadata:", results["metadatas"][idx])
    print("Content:", doc[:300], "...")  # Preview first 300 chars

# Optional: test a query
query = "What is real estate?"
query_results = collection.query(query_texts=[query], n_results=3)

print("\n🔍 Query Results:")
for i, doc in enumerate(query_results["documents"][0]):
    print(f"\nResult {i+1}:")
    print("Content:", doc[:300], "...")
    print("Score:", query_results['distances'][0][i])
```
